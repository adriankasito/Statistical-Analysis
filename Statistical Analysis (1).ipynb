```python
# Define ANSI escape codes for bold text
BOLD = '\033[1m'
RESET = '\033[0m'

# Select only numeric columns from the DataFrame
numeric_df = df.select_dtypes(include=np.number)

# Get the mean of each column 
means = np.mean(numeric_df, axis=0)
print(f"\n{BOLD}Mean of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {means[i]:.2f}")

# Get the median of each column 
medians = np.median(numeric_df, axis=0)
print(f"\n{BOLD}Median of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {medians[i]:.2f}")

# Get the standard deviation of each column 
standard_deviations = np.std(numeric_df, axis=0)
print(f"\n{BOLD}Standard Deviation of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {standard_deviations[i]:.2f}")

# Get the variance of each column
variances = np.var(numeric_df, axis=0)
print(f"\n{BOLD}Variance of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {variances[i]:.2f}")
```

```python
# Define ANSI escape codes for bold text
BOLD = '\033[1m'
RESET = '\033[0m'

# Get the mean of each column 
means = stats.tmean(numeric_df, axis=0)
print(f"\n{BOLD}Mean of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {means[i]:.2f}")

# Get the skewness of each column 
skewness = stats.skew(numeric_df, axis=0)
print(f"\n{BOLD}Skewness of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {skewness[i]:.2f}")

# Get the kurtosis of each column 
kurtosis = stats.kurtosis(numeric_df, axis=0)
print(f"\n{BOLD}Kurtosis of Each Column:{RESET}")
for i, column in enumerate(numeric_df.columns):
    print(f"{column.capitalize()}: {kurtosis[i]:.2f}")
```

Here's an explanation of skewness and kurtosis for each feature:

- **Skewness**: Skewness measures the asymmetry of the distribution of values in a column. A skewness value of 0 indicates a symmetrical distribution. Positive skewness (skewness > 0) indicates that the distribution is skewed to the right, meaning it has a long tail on the right side. Conversely, negative skewness (skewness < 0) indicates a distribution skewed to the left, with a long tail on the left side. Here:
  - Age: Skewness of 0.06 suggests a slightly right-skewed distribution.
  - BMI: Skewness of 0.28 indicates a slightly right-skewed distribution.
  - Children: Skewness of 0.94 suggests a moderately right-skewed distribution.
  - Charges: Skewness of 1.51 indicates a significantly right-skewed distribution.

- **Kurtosis**: Kurtosis measures the "tailedness" of the distribution of values in a column. A kurtosis value of 0 (mesokurtic) indicates a normal distribution, while positive kurtosis (kurtosis > 0) indicates heavier tails (leptokurtic) and negative kurtosis (kurtosis < 0) indicates lighter tails (platykurtic). Here:
  - Age: Kurtosis of -1.24 indicates a slightly platykurtic distribution.
  - BMI: Kurtosis of -0.06 suggests a distribution close to a normal distribution.
  - Children: Kurtosis of 0.20 indicates a slightly leptokurtic distribution.
  - Charges: Kurtosis of 1.60 suggests a moderately leptokurtic distribu

### Hypothesis Testing

```python
age = df['age']
bmi = df['bmi']
charges = df['charges']

# This compares the means of age between two independent groups
t_stat, p_value = stats.ttest_ind(age, bmi)
print(f"\033[1mT-statistic\033[0m: {t_stat:.4f}")
print(f"\033[1mP-Value\033[0m: {p_value:}")

# Hypothesis test result interpretation
if p_value < 0.05:
    print("Reject null hypothesis - Means of age are statistically different between the two groups")
else:
    print("Fail to reject null hypothesis - Means of age might be similar between the two groups")
```



- **T-statistic**: The T-statistic measures the difference between the means of two groups relative to the variance in the data. A higher absolute T-statistic indicates a greater difference between the means of the groups. In this case, a T-statistic of 20.4040 suggests a significant difference in the means of age between the two groups.

- **P-value**: The P-value is the probability of observing a T-statistic as extreme as, or more extreme than, the one calculated from the sample data, assuming that the null hypothesis is true. A lower P-value indicates stronger evidence against the null hypothesis. Here, the P-value is very close to zero (3.9973604269383935e-86), indicating extremely strong evidence against the null hypothesis.

- **Interpretation**: With such a low P-value, much smaller than the typical significance level of 0.05, we reject the null hypothesis. Therefore, we conclude that there is a statistically significant difference in the means of age between the two groups being compared.

- **Insights**: This result suggests that age may play a significant role or have a significant impact on whatever comparison is being made between the two groups. For example, if the two groups represent different treatment groups or populations, it indicates that age significantly influences the outcome or characteristic being studied. Further investigation into the relationship between age and the variable of interest may be warranted.

### Regression

```python
df = pd.read_csv("insurance.csv")
```

```python
# Import the libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
# Define the mappings
sex_mapping = {'male': 0, 'female': 1}
region_mapping = {'southeast': 0, 'southwest': 1, 'northeast': 2, 'northwest': 3}
smoker_mapping = {'yes': 1, 'no': 0}

# Replace string variables with integers
df['sex'].replace(sex_mapping, inplace=True)
df['region'].replace(region_mapping, inplace=True)
df['smoker'].replace(smoker_mapping, inplace=True)

df.head()
```

```python
# Define the dependent and independent variables
y = df["charges"] # dependent variable
X = df[["age", "sex", "bmi", "children", "smoker", "region"]]  # Independent variables

# Reshape the variables to fit the model
y = y.values.reshape(-1, 1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets (optional)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

# Create and fit the model
model = LinearRegression()
model.fit(X_train, y_train)

# Get the model parameters
slope = model.coef_ 
intercept = model.intercept_
r_squared = model.score(X_test, y_test)  # Calculate R-squared using test data

print("Slope (Coefficients):", slope)
print("Intercept:", intercept)
print("R-squared:", r_squared)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split, cross_val_score

# Create the linear regression model
model = LinearRegression()

# Perform cross-validation
cv_scores = cross_val_score(model, X, y, cv=5)  # 5-fold cross-validation

# Fit the model on the entire dataset
model.fit(X, y)

# Get the model parameters
slope = model.coef_ 
intercept = model.intercept_
r_squared = cv_scores.mean()  # Calculate mean R-squared score from cross-validation

# Print the model parameters
print("Slope (Coefficients):", slope)
print("Intercept:", intercept)
print("Mean R-squared (from cross-validation):", r_squared)
```

Intercept: The intercept represents the expected value of the dependent variable (charges) when all independent variables are zero. In the context of multiple linear regression (MLR), where both continuous and categorical independent variables are considered, the intercept represents the baseline value of the dependent variable when all continuous independent variables are zero and categorical variables are at their reference levels. In this case, the intercept of approximately -13036.65 indicates the expected charges when all continuous independent variables are zero and categorical variables are at their reference levels or lets say at 0.

Mean R-squared (from cross-validation): R-squared (coefficient of determination) measures the proportion of the variance in the dependent variable (charges) that is predictable from the independent variables (age, sex, bmi, children, smoker, region). The mean R-squared from cross-validation (approximately 0.747) indicates the average goodness of fit of the model across multiple cross-validation folds. A higher R-squared value suggests that the model explains a greater proportion of the variance in charges.

Insights: The slope coefficients provide insights into the relationships between the independent variables and charges, indicating the direction and magnitude of their impact. The intercept provides the baseline charges when all independent variables are zero. The mean R-squared score from cross-validation indicates the overall predictive performance of the model, suggesting that approximately 74.7% of the variance in charges is explained by the independent variables included in the model.
